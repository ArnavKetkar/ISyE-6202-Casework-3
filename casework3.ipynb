Then we are given 4-week period demand share.

```python
demand_share_4_weeks = pd.read_csv('4-week_demand_share.csv').T
demand_share_4_weeks.columns = ['mean_demand_share_4_weeks']
demand_share_4_weeks.reset_index(inplace=True)
demand_share_4_weeks.drop('index', axis=1, inplace=True)
demand_share_4_weeks
```

```python
Cv_4_week = 0.3
np.random.seed(42)

demand_share_4_weeks['sigma_4_weeks'] = Cv_4_week * demand_share_4_weeks['mean_demand_share_4_weeks']
for i in range(len(demand_share_4_weeks)):
    demand_share_4_weeks.loc[i, 'net_4_weeks_demand'] = np.random.normal(demand_share_4_weeks.loc[i, 'mean_demand_share_4_weeks'], demand_share_4_weeks.loc[i, 'sigma_4_weeks'])
demand_share_4_weeks
```

```python
demand_share_4_weeks['net_4_weeks_demand'] = demand_share_4_weeks['net_4_weeks_demand'] / demand_share_4_weeks['net_4_weeks_demand'].sum(axis=0)
demand_share_4_weeks
```

```python
daily_demand_share = pd.read_csv('demand_share_per_day.csv').T
daily_demand_share.columns = ['mean_demand_share_per_day']
daily_demand_share.reset_index(inplace=True)
daily_demand_share.drop('index', axis=1, inplace=True)
daily_demand_share
```

```python
Cv_per_day = 0.2
np.random.seed(42)

daily_demand_share['sigma_per_day'] = Cv_per_day * daily_demand_share['mean_demand_share_per_day']
for i in range(len(daily_demand_share)):
    daily_demand_share.loc[i, 'net_daily_demand'] = np.random.normal(daily_demand_share.loc[i, 'mean_demand_share_per_day'], daily_demand_share.loc[i, 'sigma_per_day'])
daily_demand_share
```

```python
daily_demand_share['net_daily_demand'] = daily_demand_share['net_daily_demand'] / daily_demand_share['net_daily_demand'].sum(axis=0)
daily_demand_share
```

```python
daily_demand_share_4_weeks = daily_demand_share.copy()
for i in range(3):
    daily_demand_share_4_weeks = pd.concat([daily_demand_share_4_weeks, daily_demand_share], ignore_index=True, axis=0)
daily_demand_share_4_weeks
```

```python
four_weekly_facs = [f for f in demand_share_4_weeks['net_4_weeks_demand'] for _ in range(28)]
len(four_weekly_facs)
```

```python
yearly_probs = pd.DataFrame({'four_weekly_facs': four_weekly_facs, 'daily_demand_share_4_weeks': list(daily_demand_share_4_weeks['net_daily_demand'])*13})
yearly_probs['yearly_prob'] = yearly_probs['four_weekly_facs'] * yearly_probs['daily_demand_share_4_weeks']
yearly_probs
```

```python
yearly_probs['yearly_prob'] = yearly_probs['yearly_prob'] / sum(yearly_probs['yearly_prob'])
yearly_probs
```

```python
annaul_demand_forecast = pd.read_csv("annual demand forecast.csv")
annaul_demand_forecast = annaul_demand_forecast.rename(columns={'1': "Year_1", '2': "Year_2", '3': "Year_3", '4': "Year_4", '5': "Year_5"})
annaul_demand_forecast
```

```python
demand_breakdown = annaul_demand_forecast[['Year_1', 'Year_2', 'Year_3', 'Year_4', 'Year_5']].copy()
demand_breakdown = demand_breakdown.loc[demand_breakdown.index.repeat(364)]
demand_breakdown.reset_index(inplace=True)
#demand_breakdown.drop('index', axis=1, inplace=True)
demand_breakdown.rename(columns={'index': 'scenario'}, inplace=True)
demand_breakdown
```

```python
for col in ['Year_1', 'Year_2', 'Year_3', 'Year_4', 'Year_5']:
    demand_breakdown[col] *= (list(yearly_probs['yearly_prob'])*5)
        
demand_breakdown
```
