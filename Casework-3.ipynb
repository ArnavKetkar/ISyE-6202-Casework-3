```python
demand_share_4_weeks = pd.read_csv('4-week_demand_share.csv').T
demand_share_4_weeks.columns = ['mean_demand_share_4_weeks']
demand_share_4_weeks.reset_index(inplace=True)
demand_share_4_weeks.drop('index', axis=1, inplace=True)
demand_share_4_weeks
```

```python
Cv_4_week = 0.3
np.random.seed(42)

demand_share_4_weeks['sigma_4_weeks'] = Cv_4_week * demand_share_4_weeks['mean_demand_share_4_weeks']
for i in range(len(demand_share_4_weeks)):
    demand_share_4_weeks.loc[i, 'net_4_weeks_demand'] = np.random.normal(demand_share_4_weeks.loc[i, 'mean_demand_share_4_weeks'], demand_share_4_weeks.loc[i, 'sigma_4_weeks'])
demand_share_4_weeks['net_4_weeks_demand'] = demand_share_4_weeks['net_4_weeks_demand']/sum(demand_share_4_weeks['net_4_weeks_demand'])
```

```python
daily_demand_share = pd.read_csv('demand_share_per_day.csv').T
daily_demand_share.columns = ['mean_demand_share_per_day']
daily_demand_share.reset_index(inplace=True)
daily_demand_share.drop('index', axis=1, inplace=True)
daily_demand_share
```

```python
Cv_per_day = 0.2
np.random.seed(42)

daily_demand_share['sigma_per_day'] = Cv_per_day * daily_demand_share['mean_demand_share_per_day']
for i in range(len(daily_demand_share)):
    daily_demand_share.loc[i, 'net_daily_demand'] = np.random.normal(daily_demand_share.loc[i, 'mean_demand_share_per_day'], daily_demand_share.loc[i, 'sigma_per_day'])
daily_demand_share
```

```python
daily_demand_share_4_weeks = daily_demand_share.copy()
for i in range(3):
    daily_demand_share_4_weeks = pd.concat([daily_demand_share_4_weeks, daily_demand_share], ignore_index=True, axis=0)
daily_demand_share_4_weeks
```

```python
def daily_simulation(day, daily_demand_df = yearly_probs, demand_df = demand_breakdown, scenario_df = df_scenario, hourly_demand_df = hourly_demand, size_df = space_probs):
    '''
    With a random probability, a scenario is chosen. Then use the scenario to choose the demand breakdown. So we get 5 years of demand breakdown.Choose a year. 


    '''
    scenario = np.random.choice(scenario_df['Scenario'].values, p=scenario_df['Probability'].values)
    year = np.random.choice([1,2,3,4,5])
    # day = np.random.choice(list(range(364)))
    
    size = np.random.choice(size_df['Space (ft3)'].values, p=size_df['Probability'].values)

    # non_zero_indices = np.nonzero(hourly_demand_df.values)
    # random_index = np.random.choice(len(non_zero_indices[0]))
    # deposit = non_zero_indices[0][random_index]
    # pickup = non_zero_indices[1][random_index]

    demand = demand_df.loc[demand_df['scenario'] == scenario, f'Year_{year}']* daily_demand_df.loc[day, 'yearly_prob']
    demand = demand.reset_index(drop=True)
    #print(demand)
    daily_demand = np.ceil(demand[day])
    #print(daily_demand)
    
    demand_yet = 0
    demand_instances = []
    while demand_yet <= daily_demand:
        # non_zero_indices = np.nonzero(hourly_demand_df.values)
        probs = np.array(hourly_demand_df).flatten()
        times = [(i,j) for i in range(24) for j in range(24)]
        (deposit, pickup) = times[np.random.choice(len(times), p = probs)]
        # random_index = np.random.choice(len(non_zero_indices[0]))
        # deposit = non_zero_indices[0][random_index]
        # pickup = non_zero_indices[1][random_index]
        size = np.random.choice(size_df['Space (ft3)'].values, p=size_df['Probability'].values)
        demand_order = np.ceil(demand[day]*hourly_demand_df.iloc[deposit, pickup])
        demand_yet += demand_order
        demand_instances.append([deposit, pickup, size,demand_order])


    return demand_instances
```

```python
import copy
def evaluation(day, n_sims = 5):
    locker_size_dict = {3: 10, 6: 10, 12: 4, 18: 3}
    locker_sizes = list(locker_size_dict.keys())
    locker_dict = {locker_size: [locker_size_dict[locker_size] for _ in range(24)] for locker_size in locker_sizes}

    rejection_rates = []
    acceptance_rates = []
    rejection_count = {locker_size: 0 for locker_size in range(1,19)}
    soujorns = {locker_size: 0 for locker_size in locker_sizes}
    utilization = {locker_size: 0 for locker_size in locker_sizes}
    for i in range(n_sims):
        locker_dict_copy = copy.deepcopy(locker_dict)
        orders = daily_simulation(day)
        orders.sort(key=lambda x: x[0])
        #print(f'Len orders = {len(orders)}')
        
        rejected_orders = []
        for order in orders:
            deposit_time = order[0]
            pickup_time = order[1]
            order_size = order[2]
            order_qty = order[3]
            # print(len(orders))

            # smallest_size_greater_than_order = [locker_size for locker_size in locker_sizes if locker_size >= order_size][0]
            # next_available_size = [locker_size for locker_size in locker_sizes if locker_size >= order_size][1] if len([locker_size for locker_size in locker_sizes if locker_size > order_size]) > 1 else None
            # second_available_size = [locker_size for locker_size in locker_sizes if locker_size >= order_size][2] if len([locker_size for locker_size in locker_sizes if locker_size > order_size]) > 2 else None
            
            eligible_lockers = [locker_size for locker_size in locker_sizes if locker_size >= order_size]

            dos = int(pickup_time-deposit_time) if pickup_time >= deposit_time else 24 - deposit_time
            for l in eligible_lockers:
                if all(locker_dict_copy[l][deposit_time:deposit_time+dos+1]) >= order_qty:
                    locker_dict_copy[l][deposit_time:deposit_time+dos] -= np.full(dos, np.ceil(order_qty))
                    soujorns[l] += dos/(n_sims*len(orders))
                    break

            else:
                rejected_orders.append(order)
                rejection_count[order_size] += 1
                #print(f'rejected order {order}')
                #break
        #print(locker_dict_copy)
            # if all(locker_dict_copy[smallest_size_greater_than_order][deposit_time:pickup_time]) >= order_qty:
            #     locker_dict_copy[smallest_size_greater_than_order][deposit_time:deposit_time+dos] -= np.full(dos, int(order_qty))
            #     soujorns[smallest_size_greater_than_order] += dos/(n_sims*len(orders))
            
            # elif next_available_size and all(locker_dict_copy[next_available_size][deposit_time:pickup_time]) >= order_qty:
            #     locker_dict_copy[next_available_size][deposit_time:deposit_time+dos] -= np.full(dos, int(order_qty))
            #     soujorns[next_available_size] += dos/(n_sims*len(orders))
            
            # elif second_available_size and all(locker_dict_copy[second_available_size][deposit_time:pickup_time]) >= order_qty:
            #     locker_dict_copy[second_available_size][deposit_time:deposit_time+dos] -= np.full(dos, int(order_qty))
            #     soujorns[second_available_size] += dos/(n_sims*len(orders))

                
        #print(f'Length rejected orders: {len(rejected_orders)}')
        rejection_rate = len(rejected_orders)/len(orders)
        rejection_rates.append(rejection_rate)
        acceptance_rate = 1 - rejection_rate
        acceptance_rates.append(acceptance_rate)
        
        for locker in locker_dict_copy.keys():
            utilization[locker] += np.mean(np.full(24, locker_dict[locker]) - locker_dict_copy[locker])/ locker_dict[locker]
            # utilization[locker] = utilization[locker][0]

    for sk in soujorns.keys():
        soujorns[sk] = np.round(soujorns[sk],2)
        
    for i in range(1,19):
        rejection_count[i] = np.round(rejection_count[i]/(n_sims),2)

    for locker in locker_dict.keys():
        utilization[locker] = np.round(utilization[locker]/n_sims,2)
        utilization[locker] = utilization[locker][0]
    
    rejection = np.round(np.mean(rejection_rates),2)
    service_level = np.round(np.mean(acceptance_rates),2)

    return rejection, service_level, utilization, soujorns, rejection_count
```

```python
rej, service, util, dos, rej_cnt = evaluation(8, n_sims=10)
print("Reject rate: ", rej)
print("Service level: ", service)
print("Utilization: ", util)
print("Avg Soujorn: ", dos)
print("Rejection count: ", rej_cnt)
```
